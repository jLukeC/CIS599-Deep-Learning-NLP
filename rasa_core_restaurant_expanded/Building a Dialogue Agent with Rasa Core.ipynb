# First step: training the NLU Model
As described in the earlier tutorial, in order to train the NLU we just need to point a Trainer at the config and data files. The model is then saved to an output directory, in this case models/nlu/current

````python
def train_nlu():
    from rasa_nlu.converters import load_data
    from rasa_nlu.config import RasaNLUConfig
    from rasa_nlu.model import Trainer

    training_data = load_data('data/franken_data.json')
    trainer = Trainer(RasaNLUConfig("nlu_model_config.json"))
    trainer.train(training_data)
    model_directory = trainer.persist('models/nlu/', fixed_model_name="current")

    return model_directory
````

# Training the Dialogue Model

````python
def train_dialogue(domain_file="restaurant_domain.yml",
                   model_path="models/dialogue",
                   training_data_file="data/babi_stories.md"):
    agent = Agent(domain_file,
                  policies=[MemoizationPolicy(), RestaurantPolicy()])

    agent.train(
            training_data_file,
            max_history=3,
            epochs=400,
            batch_size=100,
            validation_split=0.2
    )

    agent.persist(model_path)
    return agent
````

# Defining a Custom Policy with Keras
The Policy is the underlying model that Rasa Core uses to decide what action to take next based on the user's intent and the conversation context. In this case we have a small LSTM built using the Keras library.

````python
from rasa_core.policies.keras_policy import KerasPolicy

class RestaurantPolicy(KerasPolicy):
    def model_architecture(self, num_features, num_actions, max_history_len):
        """Build a Keras model and return a compiled model."""
        from keras.layers import LSTM, Activation, Masking, Dense
        from keras.models import Sequential

        n_hidden = 32  # size of hidden layer in LSTM
        # Build Model
        batch_shape = (None, max_history_len, num_features)

        model = Sequential()
        model.add(Masking(-1, batch_input_shape=batch_shape))
        model.add(LSTM(n_hidden, batch_input_shape=batch_shape))
        model.add(Dense(input_dim=n_hidden, output_dim=num_actions))
        model.add(Activation('softmax'))

        model.compile(loss='categorical_crossentropy',
                      optimizer='adam',
                      metrics=['accuracy'])

        logger.debug(model.summary())
        return model
````

# Custom Actions
This directory contains a file `yelp_search.py` that handles querying the Yelp API for information on a restaurant based on the cuisine and location. In the Action below, I've connected this API to our dialogue agent so we can incorporate the two together:

````python
import yelp_search
class RestaurantAPI(object):
    def search(self, cuisine, location, price):
        result = yelp_search.query_api(cuisine, location)
        display_location = ""
        if result:
            display_location = result["location"]["display_address"]
            display_name = result["name"]
            return "{} located at {}".format(display_name, " ".join(display_location))

        return "Nothing found!"
````

# Interactive Learning
The best feature of Rasa Core, in my opinion, is that you can train the model through interactive learning. To do so call `python bot.py train-dialogue-online`


An interactive session works as follows: you start by providing some input to the model, in this case `I'd like to eat italian food in Philadelphia`. The model will tell you the intent it detected (`inform`) the entities it found (`with cuisine:	italian, with location:	philadelphia`) along with the slots that have been filled and what action it wants to take next. From there you can correct the model's interpretations and continue the conversation.


This is also a great way to debug your model or gain insights on where it is lacking
